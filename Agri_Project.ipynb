```python
import numpy as np

# Select only numeric columns
num_cols = agri_df.select_dtypes(include=[np.number]).columns

# Replace negative values with NaN
agri_df[num_cols] = agri_df[num_cols].mask(agri_df[num_cols] < 0)
```

```python
agri_df.isnull().sum()
```

```python
# Step 1: select only numeric columns
num_cols = agri_df.select_dtypes(include="number").columns

# Step 2: loop through numeric columns
for col in num_cols:
    if agri_df[col].isna().any():   # only if NaN exists
        # Check skewness (with pandas)
        sk = agri_df[col].skew(skipna=True)
        
        if abs(sk) > 1:
            # Highly skewed → use median
            agri_df[col].fillna(agri_df[col].median(), inplace=True)
            print(f"{col} → Filled with Median")
        else:
            # Normal-ish → use mean
            agri_df[col].fillna(agri_df[col].mean(), inplace=True)
            print(f"{col} → Filled with Mean")
```

Soybean Production by Top 5 States and Yield Efficiency

```python
# Step 1: Group by state and calculate total production & area

soyabean_production_state = agri_df.groupby('state_name')['soyabean_production_tons'].sum()


soyabean_area_state = agri_df.groupby('state_name')['soyabean_area_ha'].sum()

#calculating Top 5 soyabean production state 

top_5_production = soyabean_production_state.sort_values(ascending=False).head()

top_5_area= soyabean_area_state.sort_values(ascending=False).head()

# Step 2: Calculate yield efficiency (tons per hectare) and get top 5

soyabean_yield_efficiency = top_5_production/top_5_area

top_5_yeild = soyabean_yield_efficiency[top_5_production.index]

#Creating bar chart for soyabean production state

fig, ax1 = plt.subplots(figsize=(12,7))

ax1.bar(top_5_production.index,top_5_production.values, color = 'red', edgecolor = 'black', label = 'Production',width=0.5)
ax1.set_xlabel('State',fontsize = 15)
ax1.set_ylabel('soyabean production tons',fontsize = 15 )
ax1.tick_params(axis='y', labelcolor = 'blue')

# Line chart for yield efficiency (secondary y-axis)

ax2 = ax1.twinx()

ax2.plot(top_5_production.index, top_5_yeild.values,  marker = 'o', color = 'green', linestyle = '-', linewidth = 2, label = 'yeild')
ax2.set_xlabel('State',fontsize = 15)
ax2.set_ylabel('soyabean production tons',fontsize = 0.5 )
ax2.tick_params(axis='y', labelcolor = 'blue')

# Title & legend

plt.title('Soybean Production by Top 5 States and Yield Efficiency', fontsize = 20, weight = 'bold')
plt.tight_layout()
fig.legend(loc='upper right')
plt.show()

```

Sorghum Production (Kharif and Rabi) by Region

```python
# Step 1: Group by state and calculate sorghum production (Kharif & Rabi)
sorghum_region = agri_df.groupby("state_name")[["kharif_sorghum_production_tons", "rabi_sorghum_production_tons"]].sum()

# Step 2: Take top 10 sorghum producing states for clarity
top_states = sorghum_region.sum(axis=1).sort_values(ascending=False).head(10)
sorghum_top = sorghum_region.loc[top_states.index]

# Step 3: Plot grouped bar chart
import numpy as np
fig, ax = plt.subplots(figsize=(14,7))

x = np.arange(len(sorghum_top.index))   # positions
bar_width = 0.35

bars1 = ax.bar(x - bar_width/2, sorghum_top["kharif_sorghum_production_tons"], 
               bar_width, label="Kharif Sorghum", color="orange")
bars2 = ax.bar(x + bar_width/2, sorghum_top["rabi_sorghum_production_tons"], 
               bar_width, label="Rabi Sorghum", color="green")

# Labels & Title
ax.set_xlabel("State/Region", fontsize=12)
ax.set_ylabel("Sorghum Production (tons)", fontsize=12)
ax.set_title("Sorghum Production (Kharif vs Rabi) by Region", fontsize=15, weight="bold")
ax.set_xticks(x)
ax.set_xticklabels(sorghum_top.index, rotation=45, ha="right")
ax.legend()

# Add value labels on top of bars
ax.bar_label(bars1, labels=[f"{v:.0f}" for v in sorghum_top["kharif_sorghum_production_tons"]], padding=3, fontsize=8)
ax.bar_label(bars2, labels=[f"{v:.0f}" for v in sorghum_top["rabi_sorghum_production_tons"]], padding=3, fontsize=8)

plt.tight_layout()
plt.show()
```
